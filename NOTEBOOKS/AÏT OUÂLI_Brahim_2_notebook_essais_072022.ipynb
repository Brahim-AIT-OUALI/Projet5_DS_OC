# Segmentation RFM

Fonction pour déterminer la table RFM

```python
def Table_RFM(X, Now):
    
    datetime_cols = ["order_purchase_timestamp", "order_delivered_customer_date"]
    for col in datetime_cols:
        X[col] = X[col].astype('datetime64[ns]')
    
    data_RFM = X.groupby('customer_unique_id').\
            agg({"order_purchase_timestamp": lambda x: (Now - x.max()).days,
                 "order_id": lambda x: x.nunique(),
                 "price": lambda x: sum(x)})

    data_RFM['order_purchase_timestamp'] = data_RFM['order_purchase_timestamp'].astype(int)
    data_RFM.rename(columns={'order_purchase_timestamp': 'recency',
                        'order_id': 'frequency',
                    'price': 'monetary_value'}, inplace=True)
   
    return data_RFM
    
```

#### DBSCAN

Détermination de eps et min_sample optimaux par la fonction griddb

```python
from sklearn import metrics
from sklearn.cluster import DBSCAN

eps_list = np.arange(0.1, 0.7, 0.1)
sample_list = np.arange(5,35,5)

def griddb(eps_list, sample_list, X_norm ):
    silhouette = []
    point = []
    samples = []
    
    for i in range(len(eps_list)):
        for k in range(len(sample_list)):
            
            db = DBSCAN(eps = eps_list[i], min_samples = sample_list[k] )
            db.fit(X_norm)
            silh = metrics.silhouette_score(X_norm, db.labels_)
            point.append(silh)
            if silh >= max(point):
                silhouet = silh
                sample = sample_list[k]
        silhouette.append(silhouet)
        samples.append(sample)
    
    import matplotlib.pyplot as plt
    plt.plot(eps_list, silhouette )
    
    plt.title("Silhouette score:")
    plt.xlabel("epsilon")
    plt.show()
    return(eps_list, silhouette, samples)

```
