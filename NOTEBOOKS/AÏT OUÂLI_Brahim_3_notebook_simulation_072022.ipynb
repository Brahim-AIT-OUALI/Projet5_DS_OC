#### Nous commençons en janvier 2017

```python
Date_initiale = pd.Timestamp('2017-01-01')
print(Date_initiale)
```

```python
# Pas temporel choisi pour  construire nos groupes de périodes (1ére période = 3 mois, 2emme période = 6 mois etc. )
delta_temps = pd.DateOffset(months=3)
```

```python
import pandas as pd

req_date =Date_initiale + delta_temps
print(req_date)
```

Liste des dataframes correspondants à chacune des périodes

```python
Liste_df = []
for i in range(1,8):
    Liste_df.append(df[(df['order_purchase_timestamp']>= Date_initiale) 
                       & (df['order_purchase_timestamp'] < Date_initiale + i*delta_temps)])
    
```

Détermination des tables RFM pour correspondant à chacune des périodes

```python
Liste_df[0]
```

```python
def Table_RFM(X, Now):
    
    datetime_cols = ["order_purchase_timestamp", "order_delivered_customer_date"]
    for col in datetime_cols:
        X[col] = X[col].astype('datetime64[ns]')
    
    data_RFM = X.groupby('customer_unique_id').\
            agg({"order_purchase_timestamp": lambda x: (Now - x.max()).days,
                 "order_id": lambda x: x.nunique(),
                 "price": lambda x: sum(x)})

    data_RFM['order_purchase_timestamp'] = data_RFM['order_purchase_timestamp'].astype(int)
    data_RFM.rename(columns={'order_purchase_timestamp': 'recency',
                        'order_id': 'frequency',
                    'price': 'monetary_value'}, inplace=True)
   
    return data_RFM
```

```python
Date_initiale + 1*delta_temps
```

```python
Liste_RFM = []
for i in range(1,8):
    Liste_RFM.append(Table_RFM(Liste_df[i-1], Date_initiale + i*delta_temps))
    
```

```python
Liste_RFM[0]
```

```python
def Standardisation(X):
    from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler()
    X_scaled = pd.DataFrame(scaler.fit_transform(X),index=X.index, columns = X.columns)
    return X_scaled
```

```python
liste_RFMscaled = []
for i in range(7):
    liste_RFMscaled.append(Standardisation(Liste_RFM[i]))
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=4)
```

Entrainement de notre modèle sur les données des différentes périodes

```python
liste_modele = []
for i in range (7):
    model = KMeans(n_clusters=4)
    liste_modele.append(model.fit(liste_RFMscaled[i]))
    
```

Modéle entraîné une seule fois:
Prédictions obtenues par le premier modèle (modèle entrainé sur la première période)

```python
Predictions_un_entrainement = []
for i in range(7):
    Predictions_un_entrainement.append(liste_modele[0].predict(liste_RFMscaled[i]))
    
    
```

Modéle réentraîné pour chaque période:
Pour chaque période, on refait un entraînement du modèle avant de faire les prédictions

```python
Predictions_plusieurs_entrainements = []
for i in range(7):
    Predictions_plusieurs_entrainements.append(liste_modele[i].predict(liste_RFMscaled[i]))
```

```python
liste_ARI = ["ARI1", "ARI2", "ARI3", "ARI4", "ARI5", "ARI6", "ARI7"]
Dict_ARI = {}

from sklearn.metrics import adjusted_rand_score
for i in range(7):
    Dict_ARI[liste_ARI[i]] = adjusted_rand_score(Predictions_un_entrainement[i], Predictions_plusieurs_entrainements[i])
```
